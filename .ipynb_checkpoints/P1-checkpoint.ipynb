## Helper Functions

Below are some helper functions to get started. 

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), min_line_len,max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img, lines

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

````python
#Step2: Define a kernel size and apply Gaussian smoothing
kernel_size = 5
blur_gray = gaussian_blur(gray, kernel_size)
plt.figure(figsize=(20,10))
plt.imshow(blur_gray, cmap='gray')
````

````python
#Step3: Define parameters for Canny and apply
low_threshold = 50
high_threshold = 150
edges = canny(blur_gray, low_threshold, high_threshold)
plt.figure(figsize=(20,10))
plt.imshow(edges, cmap='gray')
````

````python
#Step 4: Define region of interest and get the masked image
imshape = image.shape

mask_bottom_width = 0.85  # width of bottom edge of trapezoid, expressed as percentage of image width
mask_top_width = 0.07  # ditto for top edge of trapezoid
mask_height = 0.4  # height of the trapezoid expressed as percentage of image height
##vertices = np.array([[(150,imshape[0]),(400, 350), (490,320), (900,imshape[0])]], dtype=np.int32)
vertices = np.array([[\
        ((imshape[1] * (1 - mask_bottom_width)) // 2, imshape[0]),\
        ((imshape[1] * (1 - mask_top_width)) // 2, imshape[0] - imshape[0] * mask_height),\
        (imshape[1] - (imshape[1] * (1 - mask_top_width)) // 2, imshape[0] - imshape[0] * mask_height),\
        (imshape[1] - (imshape[1] * (1 - mask_bottom_width)) // 2, imshape[0])]]\
        , dtype=np.int32)
masked_image = region_of_interest(edges, vertices)
plt.figure(figsize=(20,10))
plt.imshow(masked_image, cmap='gray')
````

````python
#Step 5: Define the Hough Transform parameters and run it on masked_image 
#that only shows the line edges detected after masking out other edges
rho = 2 # distance resolution in pixels of the Hough grid
theta = np.pi/180 # angular resolution in radians of the Hough grid
threshold = 15    # minimum number of votes (intersections in Hough grid cell)
min_line_length = 33 #minimum number of pixels making up a line # Changed to 33 from 40
max_line_gap = 22    # maximum gap in pixels between connectable line segments # changed from 20 
#line_image = np.copy(image)*0 # creating a blank to draw lines on

# Run Hough on edge detected image
# Output "lines" is an array containing endpoints of detected line segments
line_img, lines = hough_lines(masked_image, rho, theta, threshold, min_line_length, max_line_gap)
plt.figure(figsize=(20,10))
plt.imshow(line_img)
````

````python
#step6: Overlay lane lines on the original image
lines_edges = weighted_img(line_img, image, α=0.8, β=1., γ=0.)
plt.figure(figsize=(20,10))
plt.imshow(lines_edges)
````

````python
# Function to seperate left and right lane lines based on their slope 
# slope <0 = left lane line, as the y-axis is reversed on the image
# slope > 0 = right lane line
def seperate_lef_right_lanelines(lines,img):
    left_line    = [] # (slope, intercept)
   # left_y  = [] # (length,)
    right_line   = [] # (slope, intercept)
   # right_y = [] # (length,)
    
    for line in lines:
        for x1, y1, x2, y2 in line:
            if x2==x1:
                continue # ignore a vertical line
            slope = (y2-y1)/(x2-x1)
            if slope < 0: # y is reversed in image
                if x1>img.shape[1]*0.5:
                    continue #ignore the reflectors pixels
                left_line.append([x1,y1,x2,y2])
            else:
                if x1<img.shape[1]*0.5:
                    continue #ignore the reflectors pixels
                right_line.append([x1,y1,x2,y2])
    
    return left_line,right_line # (slope, intercept), (slope, intercept)
````

````python
# Seperate x and y coordinates of lines and fit in a poly1d function 
def line_function(lines):
    x = []
    y = []
    for x1, y1, x2, y2 in lines:
        x += [x1, x2]
        y += [y1, y2]
    z = np.polyfit(x, y, 1)
    m, b = z
    #numpy has a handy function np.poly1d which can do the y = mx + b calculation for us
    f = np.poly1d(z)
    return f,x,y
````

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    global rho # distance resolution in pixels of the Hough grid
    global theta # angular resolution in radians of the Hough grid
    global threshold# minimum number of votes (intersections in Hough grid cell)
    global min_line_length#minimum number of pixels making up a line
    global max_line_gap# maximum gap in pixels between connectable line segments
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), min_line_length,max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img, lines

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)


def seperate_lef_right_lanelines(lines,img):
    left_line    = [] # (slope, intercept)
   # left_y  = [] # (length,)
    right_line   = [] # (slope, intercept)
   # right_y = [] # (length,)
    
    for line in lines:
        for x1, y1, x2, y2 in line:
            if x2==x1:
                continue # ignore a vertical line
            slope = (y2-y1)/(x2-x1)
            if slope < 0: # y is reversed in image
                if x1>img.shape[1]*0.5:
                    continue
                left_line.append([x1,y1,x2,y2])
            else:
                if x1<img.shape[1]*0.5:
                    continue #ignore the reflectors pixels
                right_line.append([x1,y1,x2,y2])
    
    return left_line,right_line # (slope, intercept), (slope, intercept)

def plot_lines(lines):
    for x1, y1, x2, y2 in lines:
        plt.plot((x1, x2), (y1, y2), 'g')
    plt.show()

def line_function(lines,framenum):
    x = []
    y = []
    for x1, y1, x2, y2 in lines:
        x += [x1, x2]
        y += [y1, y2]
        
    #A straight line can be represented with y = mx + b which is a polynomial of degree 1
    #use np.polyfit to fit a line to these points.
    #print('Frame number:',framenum)
    z = np.polyfit(x, y, 1)
    m, b = z
    #numpy has a handy function np.poly1d which can do the y = mx + b calculation for us
    f = np.poly1d(z)
    return f,x,y

def plot_merged_lines(x,f):
    for i in range(min(x), max(x)):
        plt.plot(i, f(i), 'go')

    plt.show()


# def draw_extrapolated_lanelines(image,lx,lf,rx,rf):
#     line_image = np.copy(image)*0
#     thickness = 10
#     color = (222,83,75)
#     cv2.line(image, (0, lf(0).astype(int)), (max(lx).astype(int),lf(max(lx)).astype(int)), color,thickness)
#     cv2.line(image, (min(rx), rf(min(rx)).astype(int)), (image.shape[1],rf(image.shape[1]).astype(int)), color,thickness)
#     processed_image = cv2.addWeighted(image, 0.8, line_image, 1, 0)
#     return processed_image

def draw_extrapolated_lanelines(img,mimage,lx,lf,rx,rf):
    line_image =  np.zeros((mimage.shape[0], mimage.shape[1], 3), dtype=np.uint8)
    thickness = 10
    color = (255,0,0)
    cv2.line(line_image, (0, lf(0).astype(int)), (max(lx).astype(int),lf(max(lx)).astype(int)), color,thickness)
    cv2.line(line_image, (min(rx), rf(min(rx)).astype(int)), (img.shape[1],rf(img.shape[1]).astype(int)), color,thickness)
    processed_image = cv2.addWeighted(img, 0.8, line_image, 1, 0.)
    return processed_image
````
